## üìå 1: Gather all the libraries and Data Required üîÑ

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os


# Get the absolute path of the current notebook/script
BASE_DIR = os.getcwd()

# Define the path to the data folder
DATA_DIR = os.path.join(BASE_DIR, "..", "Data")

#Load a CSV file from the Data folder
file_path = os.path.join(DATA_DIR, "cleaned_prem_data.csv")
df = pd.read_csv(file_path)

#Load league table from the Data folder
file_path2 = os.path.join(DATA_DIR, "Rankings.csv")
league_table = pd.read_csv(file_path2)
```

### üìå 2: Constructing the xG-Based League Table üî¢

In this step, we simulate an alternative **Premier League table** where results are determined **purely by expected goals (xG)** rather than actual goals. 

#### üèÜ How It Works:
- Each fixture is analyzed, and points are allocated based on **xG outcomes** rather than real match scores.
- A **+0.5 xG threshold** is used to determine winners. If a team's xG is at least **0.5 higher** than their opponent‚Äôs, they are awarded 3 points.
- Draws occur when neither team surpasses this threshold.
- The table is then ranked based on total xG-derived points.

```python
# Initialize dictionaries to hold xG stats
team_xG_points = {}
team_xG_scored = {}
team_xG_conceded = {}

# Process each fixture and allocate points and xG values
for index, row in df.iterrows():
    home_team = row['Home']
    away_team = row['Away']
    home_xg = row['Home_xg']
    away_xg = row['Away_xg']
    
    # Ensure both teams are initialized in dictionaries
    for team in [home_team, away_team]:
        if team not in team_xG_points:
            team_xG_points[team] = 0
            team_xG_scored[team] = 0
            team_xG_conceded[team] = 0

    # Allocate points based on xG comparison
    if home_xg >= away_xg + 0.5:
        team_xG_points[home_team] += 3  # Home team wins
    elif away_xg >= home_xg + 0.5:
        team_xG_points[away_team] += 3  # Away team wins
    else:
        team_xG_points[home_team] += 1  # Draw
        team_xG_points[away_team] += 1

    # Update xG scored and conceded
    team_xG_scored[home_team] += home_xg
    team_xG_conceded[home_team] += away_xg
    team_xG_scored[away_team] += away_xg
    team_xG_conceded[away_team] += home_xg

# Create a DataFrame for the xG-based league table
xG_league_table = pd.DataFrame({
    'Team': team_xG_points.keys(),
    'Points': team_xG_points.values(),
    'Scored': [team_xG_scored[team] for team in team_xG_points.keys()],
    'Conceded': [team_xG_conceded[team] for team in team_xG_points.keys()]
})


# Collect xG collumns to be rounded to integers for the table 
float_columns = ['Scored', 'Conceded']

# Round and convert only numeric columns to integers
xG_league_table[float_columns] = xG_league_table[float_columns].round(0).astype(int)

# Calculate xG Goal Difference
xG_league_table['Goal_Difference'] = xG_league_table['Scored'] - xG_league_table['Conceded']

# Rank teams based on xG points
xG_league_table = xG_league_table.sort_values(by=['Points', 'Goal_Difference'], ascending=[False, False]).reset_index(drop=True)
xG_league_table['Rank'] = xG_league_table.index + 1


# Rearrange columns
xG_league_table = xG_league_table[['Rank', 'Team',  'Points', 'Scored', 'Conceded', 'Goal_Difference']]

display(xG_league_table.style.hide(axis="index"))
```

### üìå 3: Compare the xG table with actual league table üìà

- Now we have seen the xG table, lets compare it to last seasons actual standings.
- This will highlight **ranking shifts** and **points differences** to reveal which teams overperformed or underperformed.

```python
# Rename columns for readability and merge
xG_league_table.columns = ['xG_Ranking', 'Team',  'xG_Points', 'xG_Scored', 'xG_Conceded', 'xG_Goal_Difference']

# Merge xG table with actual league table
xG_vs_actual_table = xG_league_table.merge(
    league_table[['Team', 'Ranking', 'Points']],  # Ensure correct columns from league_table
    on='Team',  # Merging on the team names
    how='left'  # Keep all xG teams and match with actual league table
)

xG_vs_actual_table = xG_vs_actual_table.sort_values(by=['Ranking'], ascending=[True]).reset_index(drop=True)

# compute ranking  differences
xG_vs_actual_table['Ranking_Change'] = xG_vs_actual_table['Ranking'] - xG_vs_actual_table['xG_Ranking']

#
avg_rank_change = xG_vs_actual_table['Ranking_Change'].abs().mean()

# Format the Ranking_Change column to show + for positive values
xG_vs_actual_table['Ranking_Change'] = xG_vs_actual_table['Ranking_Change'].apply(lambda x: f'+{x}' if x > 0 else f'{x}')


# compute  points differences
xG_vs_actual_table['Points_Change'] = xG_vs_actual_table['xG_Points'] - xG_vs_actual_table['Points']

avg_points_change = xG_vs_actual_table['Points_Change'].abs().mean()

# Format the Points_Change column to show + for positive values
xG_vs_actual_table['Points_Change'] = xG_vs_actual_table['Points_Change'].apply(lambda x: f'+{x}' if x > 0 else f'{x}')


xG_vs_actual_table.style.map(lambda x: 'font-weight: bold', subset=['Team'])


display(xG_vs_actual_table[['Team', 'Ranking', 'xG_Ranking', 'Ranking_Change', 'Points', 'xG_Points', 'Points_Change']].style.hide(axis="index"))


print(f"The average points change is: {avg_points_change:.2f}")
print(f"The average rank change is: {avg_rank_change:.2f}")
```

The API data is currently retrieved via the **Python script 'API_Current_EPL_Table.py'** found in the Additional Scripts Folder, which processes and saves the cleaned data to a CSV file for use in this notebook.

#### Step 1:  Read in the live data on EPL Standings sourced from an API and prepare the data for comparison

```python
#Load league table from the Data folder
file_path3 = os.path.join(DATA_DIR, "current_epl_standings.csv")
Current_EPL = pd.read_csv(file_path3)


#Merge the current standings with our 23/24 actual and xG table.
Rank_Comparisons = xG_vs_actual_table[['Team','Ranking','xG_Ranking']].merge(Current_EPL[['Team','Position']], on='Team')

#Rename the columns for readbility and understandability. Each value is the teams ranking in the given table
Rank_Comparisons.columns = ['Team','23/24 Table', '23/24 xG Table', 'Current 24/25 Table']

#Format for processing 
Rank_Comparisons['23/24 xG Table'] = Rank_Comparisons['23/24 xG Table'].astype('int64')

#  Calculate Overperformance in xG
Rank_Comparisons['23/24_xG_Rank_Change'] = Rank_Comparisons['23/24 Table'] - Rank_Comparisons['23/24 xG Table']

#  Calculate Rank Change from Last Season to This Season
Rank_Comparisons['This Season Rank Change'] = Rank_Comparisons['23/24 Table'] - Rank_Comparisons['Current 24/25 Table']

#  Correlation Analysis
correlation = Rank_Comparisons[['23/24_xG_Rank_Change', 'This Season Rank Change']].corr()
corr_value = correlation.loc['23/24_xG_Rank_Change', 'This Season Rank Change']
```

#### Step 2: Visualizing the Link Between xG Overperformance & Future Rank Changes üìä 

This scatter plot examines whether teams that **overperformed or underperformed** in the xG-based 23/24 league table saw **corresponding changes** in their ranking for the 24/25 season.  

#### **üîπ How It‚Äôs Built:**  
1Ô∏è‚É£ **Scatter Plot:**  
   - **X-Axis**: Change in ranking between the **actual 23/24 table** and the **xG-based 23/24 table**.  
   - **Y-Axis**: Change in ranking from **23/24 to the current 24/25 season**.  

2Ô∏è‚É£ **Regression Line (Red Line):**  
   - Highlights the overall trend between xG over/underperformance and rank changes in 24/25.  
   - A **strong correlation** would suggest that teams who overperformed in 23/24 (based on xG) tend to decline in 24/25.  

3Ô∏è‚É£ **Team Labels:**  
   - Each dot represents a **Premier League team**.  
   - **Abbreviated labels** (e.g., *MC* for Manchester City, *MU* for Manchester United) help with readability.  

4Ô∏è‚É£ **Additional Markers:**  
   - **Dashed lines at zero** indicate neutral performance‚Äîteams around (0,0) had minimal deviation from their xG expectations.  
   - **Correlation Value** is displayed to quantify the strength of the relationship.  

This visualization helps determine if **xG is predictive of future league position**, providing valuable insight into long-term performance trends. üöÄ 

```python
# Step 4: Scatter Plot with Team Labels
plt.figure(figsize=(10, 6))

# Scatter plot
sns.regplot(
    x=Rank_Comparisons['23/24_xG_Rank_Change'],
    y=Rank_Comparisons['This Season Rank Change'],
    scatter_kws={'s':100},  # Dot size
    line_kws={'color':'red'}  # Regression line color
)

# Store seen coordinates to avoid overlapping labels
seen_coords = {}

# Add team labels
for i, row in Rank_Comparisons.iterrows():
    coord = (row['23/24_xG_Rank_Change'], row['This Season Rank Change'])

    # If coordinate already exists, append the team name instead of overwriting
    if coord in seen_coords:
        seen_coords[coord] += f" & {row['Team'][:3].upper()}"
    else:
        seen_coords[coord] = row['Team'][:3].upper()

# Now, plot the labels
for coord, label in seen_coords.items():
    plt.text(
        coord[0] + 0.2,  # Slightly to the right of the point
        coord[1],  # Same Y-coordinate as the point
        label,
        fontsize=9,
        ha='left',
        va='center'
    )

# Add correlation value to the plot
plt.text(
    x=max(Rank_Comparisons['23/24_xG_Rank_Change']) * 0.8,
    y=min(Rank_Comparisons['This Season Rank Change']) * 0.9,
    s=f"Correlation: {corr_value:.2f}",
    fontsize=12,
    color='blue',
    bbox=dict(facecolor='white', edgecolor='blue', boxstyle='round,pad=0.5')
)

# Customize chart
plt.xlabel("23/24 - 23/24xG Rank Change")
plt.ylabel("23/24 - 24/25 Rank Change")


# **Fix #1: Ensure Y-axis has a clear label at 0**
plt.yticks(list(plt.yticks()[0]) + [0])

# **Fix #2: Adjust bottom of the graph to avoid large empty space**
x_min, x_max = np.floor(Rank_Comparisons['23/24_xG_Rank_Change'].min()) -1, np.ceil(Rank_Comparisons['23/24_xG_Rank_Change'].max()) +1
y_min, y_max = np.floor(Rank_Comparisons['This Season Rank Change'].min()) -1, np.ceil(Rank_Comparisons['This Season Rank Change'].max()) + 1

# Make the graph start **closer** to the lowest data point instead of a huge gap
y_min = max(y_min, min(Rank_Comparisons['This Season Rank Change']) - 1)

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

# Set tick increments
plt.xticks(np.arange(x_min, x_max + 1, 1))  # X-axis increments of 1
plt.yticks(np.arange(y_min, y_max + 2, 1))  # Y-axis increments of 2

# Add reference lines
plt.axhline(0, linestyle="--", color="gray")
plt.axvline(0, linestyle="--", color="gray")
plt.grid(True, linestyle="--", alpha=0.6)

# Show the plot
plt.show()
```
