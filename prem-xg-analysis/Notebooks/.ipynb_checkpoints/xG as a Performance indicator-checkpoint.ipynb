##  4) Chance Creation Analysis
### üîç Overview
- In this Notebook we will asses the xG both created and conceded for each team over the course of the season.
- By visualising the data through scatter charts, we will compare how teams xG for and against differs between home and away fixtures.


## üìå 1: Gather all the libraries, Data and Functions Required Libraries Data Required üîÑ

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from IPython.display import display

from Team_Name_Formatting import shorten_team_name


#read in the cleansed data developed in cleansing.py
df = pd.read_csv(r'C:\Users\d_par\OneDrive\Desktop\Danny\2025\Data Science\Portolio Projects\prem-xg-analysis\Data\cleaned_prem_data.csv')

#read in the league table put together in cleasning.py
league_table = pd.read_csv(r'C:\Users\d_par\OneDrive\Desktop\Danny\2025\Data Science\Portolio Projects\prem-xg-analysis\Data\Rankings.csv')
```

## üìå 2: Team-Level xG Creation Performance üìä
- Here, we will calculate the **average xG for** for **each team** when playing **home and away**, then visualize the results using a scatter plot.
- Each team's **league ranking** will also be displayed to highlight any **unexpectedly high or low xG values** in relation to their final league position.

### Step 1: Calculate Each Teams average xG home and away, then merge together

```python
# Calculate average xG for each team at home and away
avg_home_xg = df.groupby('Home')['Home_xg'].mean().reset_index().rename(columns={'Home': 'Team', 'Home_xg': 'Avg_Home_xg'})
avg_away_xg = df.groupby('Away')['Away_xg'].mean().reset_index().rename(columns={'Away': 'Team', 'Away_xg': 'Avg_Away_xg'})

# Merge home and away averages
avg_xg = pd.merge(avg_home_xg, avg_away_xg, on='Team', how='inner')

# Merge final league rankings
avg_xg = avg_xg.merge(league_table, on='Team', how='inner')

# Sort by Ranking
avg_xg = avg_xg.sort_values(by='Ranking', ascending=True).reset_index(drop=True)
```

### Step 2: Draw Scatter Plot 1: xG Created
This plot visualizes the relationship between **average home xG and average away xG** for each team.

```python
# Scatter Plot
plt.figure(figsize=(14, 9))
plt.scatter(avg_xg['Avg_Home_xg'], avg_xg['Avg_Away_xg'], s=100, c='blue', alpha=0.6, edgecolors='w', marker='o')


# Add text annotations with team abbreviations
for i in range(len(avg_xg)):
    short_name = shorten_team_name(avg_xg['Team'][i])
    
    # Adjust label positions to avoid overlap
    vertical_offset = -0.02 if short_name == "MU" else 0.01 if short_name == "BRI" else 0

    plt.text(
    avg_xg['Avg_Home_xg'][i] + 0.02,  
    avg_xg['Avg_Away_xg'][i] + vertical_offset,  
    f"{short_name} ({avg_xg['Ranking'][i]})",  
    fontsize=8, ha='left', va='center', fontweight='bold', color='black'
    ) 
    
# Add labels and title
plt.xlabel('Average Home xG', fontsize=12, fontweight='bold')
plt.ylabel('Average Away xG', fontsize=12, fontweight='bold')
plt.title('Scatter Plot: Home xG vs. Away xG per Team', fontsize=14, fontweight='bold', color='#333')

# Set x-axis increments to 0.2
x_min, x_max = avg_xg['Avg_Home_xg'].min(), avg_xg['Avg_Home_xg'].max()
plt.xticks(np.arange(np.floor(x_min), np.ceil(x_max) + 0.2, 0.2), fontsize=12)
plt.yticks(fontsize=12)

# Add stylish grid
plt.grid(True, linestyle='--', alpha=0.6)

# Show plot
plt.show()
```

### üîç **Key Observations **  

#### üè† **Home Dominance**  
- The **top 7 teams** all averaged **over 2 xG at home**, yet only **Manchester City** managed an average of **2 xG away from home**.  
- This strongly supports the **‚Äòhome advantage‚Äô** theory, where teams generate significantly more attacking threat in their home fixtures.  

#### üèÜ **The Makings of Champions**  
- Despite winning the league, **Manchester City** only ranked **5th** for **home xG** on average.  
- However, they had the **highest away xG of any team**, and crucially, their **home and away xG were almost identical**.  
- This highlights **City‚Äôs balanced, consistent approach**, playing **the same way home or away**, rather than relying on home advantage.  

#### üèü **Fortress Anfield: Liverpool‚Äôs Home Strength**  
- **Liverpool dominate xG at home**, generating **over 1 more xG at home than away**‚Äîmore than any other team.  
- This suggests their **attacking style flourishes significantly more at Anfield** compared to away games.  

#### ‚ö† **Relegation Trends**  
- Every team in the league managed at least **1 xG at home on average**.  
- However, the **three relegated teams** all had **below 1 xG away from home**, emphasizing their **struggles in creating chances on the road**.

## üìå 3: Team-Level xG Concession Performance üìä
- Here, we will calculate the **average xG conceded per game** for **each team** when playing **home and away**, then visualize the results using a scatter plot.
- Each team's **league ranking** will also be displayed to highlight any **unexpectedly high or low values** in relation to their final league position.

---

### Step 1:  Calculate Each Teams average xG conceded home and away, then merge together

```python
# Calculate average xG conceded (xGA) for each team at home and away
avg_home_xGA = df.groupby('Home')['Away_xg'].mean().reset_index().rename(columns={'Home': 'Team', 'Away_xg': 'Avg_Home_xGA'})
avg_away_xGA = df.groupby('Away')['Home_xg'].mean().reset_index().rename(columns={'Away': 'Team', 'Home_xg': 'Avg_Away_xGA'})

# Merge home and away xGA averages
avg_xGA = pd.merge(avg_home_xGA, avg_away_xGA, on='Team', how='inner')

# Merge final league rankings
avg_xGA = avg_xGA.merge(league_table, on='Team', how='inner')

# Sort by Ranking
avg_xGA = avg_xGA.sort_values(by='Ranking', ascending=True).reset_index(drop=True)
```

### Step 2: Draw Scatter Plot 2: xG Conceded
This plot visualizes the relationship between **average home xG conceded and average away xG conceded** for each team.

```python
import numpy as np
import matplotlib.pyplot as plt

# Scatter Plot: Relationship Between Home xGA and Away xGA for Each Team
plt.figure(figsize=(14, 9))

scatter = plt.scatter(
    avg_xGA['Avg_Home_xGA'], avg_xGA['Avg_Away_xGA'],
    s=100, c='red', alpha=0.6, edgecolors='w', marker='o'
)

# Add text annotations with specific positioning adjustments
for i in range(len(avg_xGA)):
    short_name = shorten_team_name(avg_xGA['Team'][i])  # Apply abbreviation function

    # **Default: Labels to the right**
    x_offset, y_offset = 0.015, 0  # Shift text to the right

    # **Special cases**
    if short_name in ["CRY", "WOL", "BUR"]:  
        x_offset = -0.015  # Shift text to the left
    elif short_name == "CHE":
        y_offset = -0.01  # Drop slightly downward
    elif short_name == 'BOU':
        y_offset = 0.01

    plt.text(
        avg_xGA['Avg_Home_xGA'][i] + x_offset,  
        avg_xGA['Avg_Away_xGA'][i] + y_offset,  
        f"{short_name} ({avg_xGA['Ranking'][i]})",  
        fontsize=8, ha='left' if x_offset > 0 else 'right', va='center', fontweight='bold', color='black'
    ) 

# **Set x and y axis range & increments**
x_min, x_max = 0.5, np.ceil(avg_xGA['Avg_Home_xGA'].max())  
y_min, y_max = 0.5, np.ceil(avg_xGA['Avg_Away_xGA'].max())

plt.xticks(np.arange(x_min, x_max + 0.25, 0.25), fontsize=12)
plt.yticks(np.arange(y_min, y_max + 0.25, 0.25), fontsize=12)

# Add labels and title
plt.xlabel('Average Home xGA', fontsize=12, fontweight='bold')
plt.ylabel('Average Away xGA', fontsize=12, fontweight='bold')
plt.title('Scatter Plot: Home xGA vs. Away xGA per Team', fontsize=14, fontweight='bold', color='#333')

# Add stylish grid
plt.grid(True, linestyle='--', alpha=0.6)

# Show plot
plt.show()

```

### üîç **Key Observations **   

### Defense is Key
  - The relationship between defensive stability and final league position is evident.
    
  
### üè† A Safe Home  
   - Over **half the league (12 teams) concede more than 1.75 xG away**, yet only **one team (Sheffield Utd) conceded more than 1.75 on average at home**.
   - This reinforces the notion that teams are generally more defensively sound in home fixtures.
     

### üõ° Defensive Titans
   - **Arsenal boasted the best defensive record**, conceding **the least xGA both at home and away** on average.
   - This is reflected in the final league standings, as Arsenal conceded the least goals over the 23/24 campaign.

### ‚ö† Relegation Trends  
   - Defensive weaknesses are clearly visible in the bottom three teams.  
   - **Sheffield Utd conceded by far the most xGA at home, while Luton struggled most away.**  
   - Interestingly, **Burnley, despite finishing second from bottom, displayed resilience at home**, conceding **less xGA than seven other sides‚Äîincluding 5th place Tottenham.**

## üìå 4: Summarise results with table

- Now we have visualised the data for both chance creation and concession, the below code will **summarise** the information, and display how **each teams league ranking compares with their rankings for xg created and conceded**

```python
# ‚úÖ Step 1: Merge xG Created & xG Conceded DataFrames
xg_combined = avg_xg.merge(avg_xGA, on=['Team', 'Ranking'], how='inner')

# ‚úÖ Step 2: Assign Rankings for xG and xGA
xg_combined['Rank_Home_xG'] = xg_combined['Avg_Home_xg'].rank(ascending=False, method='min')
xg_combined['Rank_Away_xG'] = xg_combined['Avg_Away_xg'].rank(ascending=False, method='min')
xg_combined['Rank_Home_xGA'] = xg_combined['Avg_Home_xGA'].rank(ascending=True, method='min')  # Lower is better
xg_combined['Rank_Away_xGA'] = xg_combined['Avg_Away_xGA'].rank(ascending=True, method='min')  # Lower is better

# ‚úÖ Step 3: Convert Rankings to Integers
xg_combined[['Rank_Home_xG', 'Rank_Away_xG', 'Rank_Home_xGA', 'Rank_Away_xGA']] = xg_combined[
    ['Rank_Home_xG', 'Rank_Away_xG', 'Rank_Home_xGA', 'Rank_Away_xGA']
].astype(int)

# ‚úÖ Step 4: Select Only Required Columns Before Styling
xg_combined = xg_combined[['Ranking', 'Team', 'Rank_Home_xG', 'Rank_Away_xG', 'Rank_Home_xGA', 'Rank_Away_xGA']]

# ‚úÖ Step 5: Sort the Table by Actual League Ranking
xg_combined = xg_combined.sort_values(by='Ranking', ascending=True).reset_index(drop=True)

# ‚úÖ Define Styling Function
def color_xg_rankings(data):
    styles = pd.DataFrame('font-size: 16px; text-align: center', index=data.index, columns=data.columns)  # Set default font size & center text
    
    for index, row in data.iterrows():
        for col in ['Rank_Home_xG', 'Rank_Away_xG', 'Rank_Home_xGA', 'Rank_Away_xGA']:
            if row[col] < row['Ranking']:
                styles.at[index, col] += '; color: green'  # Better ranking
            elif row[col] > row['Ranking']:
                styles.at[index, col] += '; color: red'  # Worse ranking
    
    # ‚úÖ Make team names bold
    styles['Team'] = 'font-weight: bold; font-size: 16px; text-align: left'
    
    return styles

# ‚úÖ Apply the function to style the table
styled_xg_combined = xg_combined.style.apply(color_xg_rankings, axis=None)

# ‚úÖ Center-align column headers
styled_xg_combined.set_table_styles([
    {'selector': 'th', 'props': [('text-align', 'center')]}
])

# ‚úÖ Display the Final Table Without Index
display(styled_xg_combined.hide(axis="index"))
```
